## Task 1: Basic Monte Carlo Integration (Uniform Sampling)

### Problem:
Estimate the integral of $f(x) = x^2$ over the interval $[0, 1]$ using Monte Carlo integration.

### Instructions:
1. Implement a Monte Carlo simulation with uniform sampling.
2. Compare the result with the analytical integral value.
3. Visualize the average block areas using random points.


```python

import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x ** 2

# Number of samples
N = 1000

# Generate N random samples in [0, 1]
x_samples = np.random.uniform(0, 1, N)

# Monte Carlo estimate of the integral
integral_estimate = np.mean(f(x_samples))

# Analytical value of the integral
analytical_value = 1/3

# Output the results
print(f"Monte Carlo Estimate: {integral_estimate}")
print(f"Analytical Value: {analytical_value}")

# Visualization
plt.figure(figsize=(8, 5))
plt.bar(x_samples, f(x_samples), width=1/N, color='blue', alpha=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Average Block Areas for f(x) = x^2")
plt.show()
```


## Task 2: Midpoint Rule with Random Sub-Intervals

### Problem:
Estimate the integral of $f(x) = \sin(x)$ over $[0, \pi]$ using the random midpoint rule.

### Instructions:
1. Randomly sample midpoints within sub-intervals.
2. Compare with the analytical result.
3. Visualize the process.


```python

# Define the function
def f(x):
    return np.sin(x)

# Integration interval
a, b = 0, np.pi

# Number of samples
N = 1000

# Generate random midpoints
midpoints = np.random.uniform(a, b, N)

# Monte Carlo estimate using midpoints
integral_estimate = (b - a) * np.mean(f(midpoints))

# Analytical value of the integral
analytical_value = 2

# Output the results
print(f"Monte Carlo Estimate: {integral_estimate}")
print(f"Analytical Value: {analytical_value}")

# Visualization
plt.figure(figsize=(8, 5))
plt.hist(midpoints, bins=30, alpha=0.6, color='orange', edgecolor='black')
plt.title("Random Midpoints for sin(x) Integration")
plt.show()
```
